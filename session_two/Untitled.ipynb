## Michaelis–Menten kinetics

$$rate =    \frac{d[P]}{dt}                  = \frac{V_{max}[S]}{K_{M}+[S]} $$

```python


#Label and isolate the raw data
conc = np.array([1.0,2.0,5.0,10.0, 20.0, 40.0, 100.0, 200.0])
rate = np.array([0.02,0.06,0.13,0.2,0.31,0.47,0.59,0.71])

#Michaelis-Menten Kinetics
def rate_opt(c, km, vmax):
    return (c*vmax)/(km+c)

g = [29.5,0.79]
n = len(conc)
y = np.empty(n)
for i in range(n):
    y[i] = rate_opt(conc[i], g[0], g[1])
    

#Plot the data, curve, and formatting
plt.figure(figsize=(5,5))

plt.plot(conc,y,lw=3)

plt.xlabel('Concentration / mol dm$^{-3}$',fontsize=18) 
plt.ylabel('Rate',fontsize=18)
plt.title('Michaelis–Menten kinetics',pad=12, fontsize=25)
plt.grid(linestyle='dotted',c='0.7')
plt.plot(conc,rate,lw=1,c='#2909B9',label='"Curve"')
plt.scatter(conc,rate,s=30,c='#A16DD8',alpha=0.6,label='Raw Data')
plt.tick_params(color='0.9',labelsize=18)
plt.legend(frameon=False,fontsize=16)
plt.minorticks_on()
plt.show()
```

```python
from scipy.optimize import curve_fit

k,cov = curve_fit(rate_opt,conc,rate, g)
print(k)


y = np.empty(n)
for i in range(n):
    y[i] = rate_opt(conc[i], k[0], k[1])
    
def rate_smooth(s,km,vmax):
    return (s*vmax)/(km+s)

base = np.linspace(0, 200, 200)
p = np.empty(200)
p_len = len(p)
for i in range(p_len):
    p[i] = rate_smooth(base[i],k[0],k[1])
    
    
#Plot the data, curve, and formatting
plt.figure(figsize=(12,8))

plt.plot(conc,rate,lw=0.25,c='#2909B9',label='"Curve"')
plt.plot(conc,y,lw=0.5)
plt.plot(base,p,lw=3,label='"Test"')
plt.scatter(conc,rate,s=30,c='#A16DD8',alpha=0.6,label='Raw Data')

plt.xlabel('Concentration / mol dm$^{-3}$',fontsize=18) 
plt.ylabel('Rate',fontsize=18)
plt.title('Michaelis–Menten kinetics',pad=12, fontsize=25)
plt.grid(linestyle='dotted',c='0.7')
plt.tick_params(color='0.9',labelsize=18)
plt.legend(frameon=False,fontsize=16)
plt.minorticks_on()
plt.show()
```

```python
print(r2_score(rate,y))
```
